```python
import plotly.express as px
deaths = pd.read_csv("total-number-of-deaths-by-cause.csv")
deaths.loc[deaths['Deaths - Non-communicable diseases - Sex: Both - Age: All Ages (Number)']< 2.e6,
 'Entity'] = 'Other countries' # Represent only large countries
deaths = px.pie(deaths, values='Deaths - Non-communicable diseases - Sex: Both - Age: All Ages (Number)',
 names='Entity', title='Total number of deaths caused by Non-communicable Diseases')
deaths.show()
```
